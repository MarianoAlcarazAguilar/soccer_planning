# Gr√°fica del progreso

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('./progreso_calificacion.csv')
```

```python
fig, ax = plt.subplots(figsize=(15,5))
limite_superior = .8

ax.spines.top.set_visible(False)
ax.spines.right.set_visible(False)

ax.plot(datos.Calificacion, zorder=3)
ax.hlines(limite_superior, 0, datos.Calificacion.size, color='gray', zorder=1)
ax.set_title('Progreso de Calificaci√≥n', size=18, fontname='Arial')
ax.set_xlabel('N√∫mero de Iteraci√≥n', size=13, fontname='Arial')
ax.set_ylabel('Calificaci√≥n', size=13, fontname='Arial')

fig.tight_layout()
# fig.savefig('./progreso_calificacion.jpeg', dpi=300)

plt.show()
```
